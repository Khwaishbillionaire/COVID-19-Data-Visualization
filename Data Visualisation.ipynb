```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data=r'C:\Users\Khwaish Goel\Downloads\archive\country_wise_latest.csv'
```

```python
df=pd.read_csv(data)
```

```python
df.info()
```

```python
#sorting data frame by confirmed cases and selecting top 5 countries
```

```python
top_countries=df.sort_values(by='Confirmed', ascending=False).head(5)
```

```python
top_countries
```

```python
#creating a bar using matplot
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.bar(top_countries['Country/Region'],top_countries['Confirmed last week'],color='red')
plt.xlabel('Countries')
plt.ylabel('No.of confirmed cases')
```

```python
# Plotting the lines for Recovered and Deaths with countries on the x-axis
```

```python
plt.figure(figsize=(18, 8))  # Larger figure size
plt.plot(top_countries['Country/Region'], top_countries['Recovered'], label='Recovered', color='green', marker='o')
plt.plot(top_countries['Country/Region'], top_countries['Deaths'], label='Deaths', color='red', marker='x')
plt.title('COVID-19 Recovered vs Deaths by Country')
plt.xlabel('Country')
plt.ylabel('Count')

```
